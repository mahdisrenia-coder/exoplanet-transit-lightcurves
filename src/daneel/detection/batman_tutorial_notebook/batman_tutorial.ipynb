```python
import batman
import numpy as np
import matplotlib.pyplot as plt
np.__version__
```

```python
limb_dark_data = np.genfromtxt("/Users/tiziano/Dropbox/Computational Astrophysics/24_25/notebooks/HandsOn Lesson notebooks/ExoCTK_G0_results.txt", skip_header=2)
u1 = np.mean(limb_dark_data[:, 8])
u2 = np.mean(limb_dark_data[:, 10])
print(u1, u2)
```

```python
params = batman.TransitParams()
params.t0   = 0.                #time of inferior conjunction
params.per  = 2.21857567        #orbital period in days
params.rp   = 0.1548291121      #planet radius (in units of stellar radii)
params.a    = 8.9625402483      #semi-major axis (in units of stellar radii)
params.inc  = 85.71             #orbital inclination (in degrees)
params.ecc  = 0.                #eccentricity
params.w    = 20.               #longitude of periastron (in degrees)
params.u    = [u1, u2]          #limb darkening coefficients [u1, u2]
params.limb_dark = "quadratic"  #limb darkening model
```

```python
t = np.linspace(-0.075, 0.075, 1000)
```

```python
m = batman.TransitModel(params, t)	        #initializes model
flux = m.light_curve(params)
```

```python
plt.plot(t, flux)
plt.xlabel("Time from central transit (days)")
plt.ylabel("Relative flux")
# plt.ylim((0.989, 1.001))
plt.savefig("lc.png") 
plt.show()
```
